```python
# Install KeyBERT (Uncomment the following line if you haven't installed it yet)


from keybert import KeyBERT

def extract_keywords(text):
    # Initialize KeyBERT with the BERT model
    kw_model = KeyBERT()

    # Extract keywords with their relevance scores
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 2), stop_words='english')

    # Return the keywords with scores
    return keywords

def main():
    # Example text
    text = "Deep learning models are effective for keyword extraction in various natural language processing tasks."

    # Extract keywords
    keywords = extract_keywords(text)

    # Display the keywords with scores
    for keyword, score in keywords:
        print(f"{keyword}: {score:.4f}")

if __name__ == "__main__":
    main()
```

```python
from rapidfuzz import process, fuzz

# Preloaded dictionary of words
preloaded_words = {
    "apple",
    "banana",
    "grape",
    "orange",
    "watermelon",
    'cancel today',
    'fee cancel'
}

# Function to find the best match
def find_best_match(input_word, preloaded_words, threshold=90):
    # Find the best match with a score
    best_match = process.extractOne(input_word, preloaded_words, scorer=fuzz.partial_ratio)

    # best_match is a tuple (match, score, index)
    if best_match is not None:
        match, score, _ = best_match
        # If the best match score is above the threshold, return the best match
        if score >= threshold:
            return match

    # If no match is found, add the input word to the dictionary and return it
    preloaded_words.add(input_word)
    return input_word

# Example usage
input_word = 'cancelation'  # Example input word
result = find_best_match(input_word, preloaded_words)
print(f"Output word: {result}")
print(f"Updated dictionary: {preloaded_words}")

```

```python
from datetime import datetime

def calculate_reply_delay_in_minutes(sent_time_str, reply_time_str):
    # Define the format of the date and time strings
    time_format = "%d-%m-%Y %H:%M:%S"

    # Parse the sent and reply time strings into datetime objects
    sent_time = datetime.strptime(sent_time_str, time_format)
    reply_time = datetime.strptime(reply_time_str, time_format)

    # Calculate the delay
    delay = reply_time - sent_time

    # Convert the delay to minutes
    delay_in_minutes = delay.total_seconds() / 60

    return delay_in_minutes

# Example usage
sent_time_str = "31-07-2024 14:16:30"  # Sent time
reply_time_str = "1-08-2024 15:30:45"  # Reply time

delay_minutes = calculate_reply_delay_in_minutes(sent_time_str, reply_time_str)
print(f"Reply delay: {delay_minutes:.2f} minutes")

```

```python
#code for sentiment analysis fucntion
import torch
import pandas as pd
import numpy as np
import rapidfuzz

from scipy.special import softmax
import transformers
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification

MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

def sentimentscore(example):
    encoded_text=tokenizer(example,return_tensors='pt')
    output=model(**encoded_text)
    scores=output[0][0].detach().numpy()
    scores=softmax(scores)
    final_score= (scores[0]*(-1)) +scores[2]
    if final_score>=0.5:
        return 1
    elif final_score<-0.5:
        return -1
    else:
        return 0


#Code for keyword extraction function setup


from keybert import KeyBERT

def extract_keywords(text):
    # Initialize KeyBERT with the BERT model
    kw_model = KeyBERT()

    # Extract keywords with their relevance scores
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 2), stop_words='english')

    # Return the keywords with scores
    return keywords

def main():
    # Example text
    text = "Deep learning models are effective for keyword extraction in various natural language processing tasks."

    # Extract keywords
    keywords = extract_keywords(text)

    # Display the keywords with scores
    for keyword, score in keywords:
        print(f"{keyword}: {score:.4f}")

if __name__ == "__main__":
    main()

#code for repeated keyword cancelation

from rapidfuzz import process, fuzz

# Preloaded dictionary of words
preloaded_words = {
    "apple",
    "banana",
    "grape",
    "orange",
    "watermelon",
    'cancel today',
    'fee cancel'
}

# Function to find the best match
def find_best_match(input_word, preloaded_words, threshold=90):
    # Find the best match with a score
    best_match = process.extractOne(input_word, preloaded_words, scorer=fuzz.partial_ratio)

    # best_match is a tuple (match, score, index)
    if best_match is not None:
        match, score, _ = best_match
        # If the best match score is above the threshold, return the best match
        if score >= threshold:
            return match

    # If no match is found, add the input word to the dictionary and return it
    preloaded_words.add(input_word)
    return input_word

# Example usage
input_word = 'cancelation'  # Example input word
result = find_best_match(input_word, preloaded_words)
print(f"Output word: {result}")
print(f"Updated dictionary: {preloaded_words}")


#Code for date time comparision

from datetime import datetime

def calculate_reply_delay_in_minutes(sent_time_str, reply_time_str):
    # Define the format of the date and time strings
    time_format = "%d-%m-%Y %H:%M:%S"

    # Parse the sent and reply time strings into datetime objects
    sent_time = datetime.strptime(sent_time_str, time_format)
    reply_time = datetime.strptime(reply_time_str, time_format)

    # Calculate the delay
    delay = reply_time - sent_time

    # Convert the delay to minutes
    delay_in_minutes = delay.total_seconds() / 60

    return delay_in_minutes

# Example usage
sent_time_str = "31-07-2024 14:16:30"  # Sent time
reply_time_str = "1-08-2024 15:30:45"  # Reply time

delay_minutes = calculate_reply_delay_in_minutes(sent_time_str, reply_time_str)
print(f"Reply delay: {delay_minutes:.2f} minutes")

#code for scraping and SQL update
from google_play_scraper import reviews, Sort
import pandas as pd
import mysql.connector
from datetime import datetime

# Define the app package name for Ola
app_package_name = 'com.olacabs.customer'

# Define the limit for the number of reviews to fetch
review_limit = 10000  # Increase the limit if needed
fetched_reviews = 0

preloaded_words = {
    "apple",
    "banana",
    "grape",
    "orange",
    "watermelon",
    'cancel today',
    'fee cancel'
}

def truncate_text(text, max_length=512):
    """Truncate text to the maximum length supported by the model."""
    if len(text) > max_length:
        return text[:max_length]
    return text

def calculate_reply_delay_in_minutes(at, repliedAt):
    """Calculate the delay in minutes between the review and the reply."""
    delay = (repliedAt - at).total_seconds() / 60
    return int(delay)

def process_review(reviewId, userName, userImage, content, score, thumbsUpCount,
                   reviewCreatedVersion, at, replyContent, repliedAt, appVersion):
    """Process each review and handle possible errors."""
    try:
        content = truncate_text(content)  # Ensure content length is manageable
        P_N_N = sentimentscore(content)
        keywords = extract_keywords(content)

        at_str = at.strftime('%Y-%m-%d %H:%M:%S')  # Convert datetime to string
        if replyContent is None:
            r_status = 0
            d = None
        else:
            r_status = 1
            repliedAt_str = repliedAt.strftime('%Y-%m-%d %H:%M:%S')  # Convert datetime to string
            d = calculate_reply_delay_in_minutes(at, repliedAt)
            if d < 0:
              d=0
              r_status=0
            else:
              d=d

        dist_nk = []
        for word in keywords:
            nk = find_best_match(word[0], preloaded_words, threshold=90)
            if nk not in dist_nk:
                dist_nk.append(nk)
                new_row = {
                    'Keyword': nk, 'Content': content, 'sentiment_score': P_N_N,
                    'Date_of_posting': at_str, 'Reply_status': r_status, 'reply_delay': d,
                    'Key_relevance': word[1]
                }
                # Insert new_row into SQL database
                cursor.execute('''INSERT INTO processed_reviews (Keyword, Content, sentiment_score, Date_of_posting, Reply_status, reply_delay, Key_relevance)
                                  VALUES (%s, %s, %s, %s, %s, %s, %s)''',
                               (new_row['Keyword'], new_row['Content'], new_row['sentiment_score'], new_row['Date_of_posting'],
                                new_row['Reply_status'], new_row['reply_delay'], new_row['Key_relevance']))
    except Exception as e:
        print(f"Error processing review {reviewId}: {e}")

# Connect to MySQL database
conn = mysql.connector.connect(
    host='localhost',  # e.g., 'localhost'
    user='root',
    password='Csbawaskar@123',
    database='ReviewsDB'  # e.g., 'ReviewsDB'
)
cursor = conn.cursor()

# Create table if it doesn't exist
cursor.execute('''CREATE TABLE IF NOT EXISTS processed_reviews (
                  Keyword VARCHAR(255),
                  Content TEXT,
                  sentiment_score FLOAT,
                  Date_of_posting DATETIME,
                  Reply_status INT,
                  reply_delay INT,
                  Key_relevance FLOAT)''')

# Fetch reviews
result, continuation_token = reviews(
    app_package_name,
    lang='en',  # Language of reviews
    country='IN',  # Country of reviews
    count=review_limit,  # Number of reviews to fetch
    sort=Sort.NEWEST  # Sorting the reviews by newest
)

# Process each review
for review in result:
    process_review(
        review['reviewId'], review['userName'], review['userImage'], review['content'],
        review['score'], review['thumbsUpCount'], review['reviewCreatedVersion'],
        review['at'], review['replyContent'], review['repliedAt'], review['appVersion']
    )

fetched_reviews += len(result)
print(f"Fetched {fetched_reviews} reviews.")

# Continue fetching more reviews if needed
while continuation_token and fetched_reviews < review_limit:
    print(f"Continuing with token: {continuation_token}")
    additional_reviews, continuation_token = reviews(
        app_package_name,
        lang='en',  # Language of reviews
        country='IN',  # Country of reviews
        count=review_limit,  # Number of reviews to fetch
        sort=Sort.NEWEST,  # Sorting the reviews by newest
        continuation_token=continuation_token  # Use the continuation token to fetch more reviews
    )

    for review in additional_reviews:
        process_review(
            review['reviewId'], review['userName'], review['userImage'], review['content'],
            review['score'], review['thumbsUpCount'], review['reviewCreatedVersion'],
            review['at'], review['replyContent'], review['repliedAt'], review['appVersion']
        )

    fetched_reviews += len(additional_reviews)
    print(f"Fetched {fetched_reviews} reviews.")

# Commit and close connection to save all inserted rows
conn.commit()
conn.close()
```
